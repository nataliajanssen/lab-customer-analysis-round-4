```python
#### cleaning Income
len(data_numerical[data_numerical['Income']<=0]) 

data_numerical['Income'] =np.where(data_numerical['Income']<0,0,data_numerical['Income'])


mean= np.sum(data_numerical['Income'])/len(data[data_numerical['Income']>0])

data_numerical['Income']=data_numerical['Income'].replace(0,mean)

####cleaning Months Since Last Claim
len(data_numerical[data_numerical['Months Since Last Claim']<=0]) 

data_numerical['Months Since Last Claim'] =np.where(data_numerical['Months Since Last Claim']<0,0,data_numerical['Months Since Last Claim'])


mean= np.sum(data_numerical['Months Since Last Claim'])/len(data[data_numerical['Months Since Last Claim']>0])

data_numerical['Months Since Last Claim']=data_numerical['Months Since Last Claim'].replace(0,mean)

####cleaning Months Since Policy inception

len(data_numerical[data_numerical['Months Since Policy Inception']<=0]) 

data_numerical['Months Since Policy Inception'] =np.where(data_numerical['Months Since Policy Inception']<0,0,data_numerical['Months Since Last Claim'])


mean= np.sum(data_numerical['Months Since Policy Inception'])/len(data[data_numerical['Months Since Policy Inception']>0])

data_numerical['Months Since Policy Inception']=data_numerical['Months Since Last Claim'].replace(0,mean)

####cleaning Number of Open Complaints

len(data_numerical[data_numerical['Number of Open Complaints']<=0]) 

data_numerical['Number of Open Complaints'] =np.where(data_numerical['Number of Open Complaints']<0,0,data_numerical['Number of Open Complaints'])


mean= np.sum(data_numerical['Number of Open Complaints'])/len(data[data_numerical['Number of Open Complaints']>0])

data_numerical['Number of Open Complaints']=data_numerical['Number of Open Complaints'].replace(0,mean)
```
